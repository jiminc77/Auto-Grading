## Problem 2: Integral Transforms (Laplace Transform)

### Statement
A semi-infinite domain ($x > 0$) initially at $T_0 = 25$ K experiences a step heating at $x=0$:
- Boundary condition: $T(0, t) = 100$ K for $t > 0$
- Initial condition: $T(x, 0) = 25$ K
- Thermal diffusivity: $\alpha = 2 \times 10^{-5}$ m²/s

### Tasks
1. **Solve using Laplace Transforms:**
   - Apply the Laplace transform to the heat equation in $t$: $\mathcal{L}\{\frac{\partial T}{\partial t}\} = s\bar{T}(x,s) - T(x,0)$

   - Invert the Laplace transform using the complementary error function: $\text{erfc}(\eta) = 1 - \text{erf}(\eta)$

2. **Expected result:**
