## Problem 3: Green's Functions and Duhamel's Principle

### Statement
A 1D domain with a volumetric heat source $S(x, t)$ is governed by:
$$\frac{\partial T}{\partial t} = \alpha \frac{\partial^2 T}{\partial x^2} + S(x,t)$$

Consider:
- Domain: $x \in [0, L]$ with $L = 1$ m
- Initial condition: $T(x, 0) = 0$ K
- Boundary conditions: $T(0, t) = T(L, t) = 0$ K (Dirichlet)
- Heat source: $S(x, t) = S_0 \cdot \sin(\pi x / L) \cdot e^{-t/\tau}$ (localized, decaying)
  - $S_0 = 10$ W/m
  - $\tau = 10$ s

### Tasks
1. **Explain Duhamel's Principle:**
   - Why can we treat a time-varying source as a superposition of impulse responses?
   - Write the general form: $T(x,t) = \int_0^t \int_0^L G(x, t; \xi, \tau) S(\xi, \tau) d\xi d\tau$

2. **Compute the Green's function:**
   - For Dirichlet BCs on a finite slab, the Green's function (eigenfunction expansion) is:
   $$G(x, t; \xi, \tau) = \frac{2}{L} \sum_{n=1}^{\infty} \sin(n\pi x/L) \sin(n\pi \xi/L) \exp[-\alpha(n\pi/L)^2(t-\tau)]$$

3. **Implement and analyze:**
   - Compute $T(x, t)$ using Duhamel's principle (use $N=10$ eigenmodes)
   - Plot the solution at selected times
   - Compare: how does the temperature evolve as the source decays?

```python
from scipy.integrate import dblquad

# Green's function for Dirichlet BC on finite slab
def greens_function(x, t, xi, tau, L=1.0, alpha=1e-5, N=10):
    """Finite slab Green's function (eigenfunction expansion)."""
    if t <= tau:
        return 0
    G = 0
    for n in range(1, N+1):
        lambda_n = n * np.pi / L
        G += (2/L) * np.sin(lambda_n * x) * np.sin(lambda_n * xi) * np.exp(-alpha * lambda_n**2 * (t - tau))
    return G

# Heat source
def source(x, t, S0=10, tau=10, L=1.0):
    return S0 * np.sin(np.pi * x / L) * np.exp(-t / tau)

# Duhamel's principle solution
def duhamel_solution(x, t, L=1.0, alpha=1e-5):
    """Compute T(x,t) using Duhamel's principle."""
    def integrand(tau, xi):
        return greens_function(x, t, xi, tau, L, alpha) * source(xi, tau, L=L)

    # Double integral over tau and xi
    result, _ = dblquad(integrand, 0, L, 0, t)
    return result

# Compute and plot the solution
# YOUR CODE HERE
```
